```python
import pandas as pd
import numpy as np
import ast
```

```python
creadit=pd.read_csv('tmdb_5000_credits.csv')
movie=pd.read_csv('tmdb_5000_movies.csv')
```

```python
creadit.head()
```

```python
movie.head()
```

```python
movie=pd.merge(movie,creadit,on='title')
```

```python
movie.head()
```

```python
movie.info()
```

```python
movie=movie.loc[ :, ['genres','movie_id','keywords','title','overview','cast','crew']]
```

```python
movie.info()
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie.duplicated().sum()
```

```python
movie.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# covert ["Action","Adventure","Fantasy","science fiction"]
def convert_genre(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movie['genres']=movie['genres'].apply(convert_genre)
```

```python
movie['keywords']=movie['keywords'].apply(convert_genre)
```

```python

#'[{"cast_id": 242, "character": "Jake Sully", "credit_id": "5602a8a7c3a3685532001c9a", "gender": 2, "id": 65731, "name": "Sam Worthington", "order": 0}, {"cast_id": 3, "character": "Neytiri", "credit_id": "52fe48009251416c750ac9cb", "gender": 1, "id": 8691, "name": "Zoe Saldana", "order": 1},
#[name_actor,,]
def convert_cast(obj):
    list=[]
    counter=0
    for i in ast.literal_eval(obj):
        
        if counter !=3:
            list.append(i['name'])
            counter+=1
        else :
            break
    return list
```

```python
movie['cast']=movie['cast'].apply(convert_cast)
```

```python
#'[{"credit_id": "52fe48009251416c750aca23", "department": "Editing", "gender": 0, "id": 1721, "job": "Editor", "name": "Stephen E. Rivkin"}
def convert_crew(obj): #director find
    list=[]
    counter=0
    for i in ast.literal_eval(obj):
        
        if i['job']=='Director':
            list.append(i['name'])
            break
            
    return list
```

```python
movie['crew']=movie['crew'].apply(convert_crew)
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split()) #overview string list to concatinate
```

```python
movie.head()
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie['tags']=movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']
```

```python
movie.head()
```

```python
new_df=movie[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie_1):
    movie_index=new_df[new_df['title']==movie_1].index[0] #movie index find
    distances=similarity[movie_index] #index loss after sort
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    #movie_list index similarity
    for i in movies_list:
        print(new_df.iloc[i[0]].title) #index find title
```

```python
new_df[new_df['title']=='Batman Begins'].index[0]
```

```python
recommend('Batman Begins')
```
